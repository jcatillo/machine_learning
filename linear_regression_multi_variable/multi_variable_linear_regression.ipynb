Install necesarry modules/libraries

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("./train_data/homeprices.csv") #add dataframe
df
```

See that there is a NaN, meaning there is no data for series 3's bedroom. In order to fix that, i'll use the median of the bedrooms and plug it in series 3's number of bedroom

```python
import math #math module from python
median_bedrooms = math.floor(df.bedrooms.median()) #this takes the median of the bedrooms among the series. math.floor is used to round it down to the nearest whole number (integer)
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms) #this fills in any NaN values under bedrooms feature with the median of the bedrooms which in this case is 3
df
```

This is one of the method to preprocess or clean your data. Before you train the model, you should have a clean data so that errors will not occur. Hence, we filled in the NaN values with the 
median of the bedrooms which is 3.5, and since there is no such thing as a .5 bedroom, we round it down to 3.0

```python
#create linear regression model
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']], df.price) #fit method is used to train the model
#parameter 1: independent variables : area, bedroom, age
#parameter 2: dependent variable: price
```

```python
reg.coef_
```

```python
reg.intercept_
```

if we go back, in order to predict the price, this formula is used:
price = m1 * area + m2 * bedrooms + m3 * age + b
the "m"s are also called as weights(slope)/coefficients. the b is also called as bias(y-intercept)
if we plug in the reg.coef_ the following are the values:
m1 = 137.25
m2 = -26025.0
m3 = -6825.0
b(y-intercept) = 383724.99999999994

and since we have our variables, we can not predict/calculate the price

predict the following:
3000 sqft, 3 bedrooms, 40 yrs old
2500 sqft, 4 bedrooms, 5 yrs old

```python
X_new = pd.DataFrame([[3000, 3, 40]], columns=["area", "bedrooms", "age"])

price = reg.predict(X_new)
print(f"price: {price}")
```

If we compare it to the given data that we trained the model with:
