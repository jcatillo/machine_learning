---------- Training phase --------------

Import necessary modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("./train_data/homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area in sqft')
plt.ylabel('price(USD)')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
plt.xlabel('area in sqft')
plt.ylabel('price(USD)')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
plt.show()
```

```python
reg.predict([[5000]]) #Y'
```

```python
reg.coef_ #weight
```

```python
reg.intercept_ #bias
```

```python
m = reg.coef_ #weight / slope
b = reg.intercept_ #bias / y-intercept
x = 5000 # the thing that we want to predict

y = m*x+b #Linear function (linear equation)

print(y) #The prediction
```

---------------- Testing phase ----------------

```python
d = pd.read_csv("./test_data/homeprices.csv")
d.head(3) # shows the first 3 rows
```

```python
p = reg.predict(d) # predicts the prices of the houses with respect to the area
```
